## Examples of Fourier series



```python
## preamble

# use this for static plotting, including 3D plots  :
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt
```

## First example: a trapezoidal function

```python
def trapezoid(x):
    return np.piecewise(x, [(x >=0) & (x < 1), (x >= 1) & (x < 2), (x >= 2) & (x < 3)], [lambda x: x, lambda x: 1, lambda x: 3-x])

L = 3
```

Let's plot three periods. Note that I only defined the function on [0,L] so the plot will show 0 outside .

```python
x = np.linspace(-3, 6, 700) 
y = trapezoid(x)
plt.plot(x,y)
plt.show()
```

Let's make it look periodic by adding copies shifted by the period, in both directions

```python
def per_trapezoid(x, num):
    value = trapezoid(x) # initialize with the unshifted value
    for n in range(1, num + 1):
        value += trapezoid(x - n * L) + trapezoid(x + n * L)
    return value

num = 6

x = np.linspace(-3, 6, 700) 
y = per_trapezoid(x, num)
plt.plot(x,y)
plt.show()
```

### Sine and cosine series.
We want to write the series in the form $f(x) = a_0 + \sum_{n=1}^{\infty} \left(a_n \cos {2n \pi x \over L} +  b_n \sin {2n \pi x \over L} \right)$ where the coefficients are given by the usual formulae:
$$ a_0 = {1\over L}\int_{0}^{L}f(x)$$
and for all $n\ge 1$:
$$a_n = {2\over L}\int_{0}^{L}f(x)\cos{2 n\pi x\over L}$$
$$b_n = {2\over L}\int_{0}^{L}f(x)\sin{2 n\pi x\over L}$$


I will evaluate them numerically below for this trapezoidal function. If you use scipy, you can get the symbolic results.


```python
# Number of terms in the Fourier series
num_terms = 50

# Generate x values for one period
x_values = np.linspace(0, L, 600)

# Calculate the Fourier series coefficients
a0 = (1 / L) * np.trapz(trapezoid(x_values), x_values)  # get a0 using numerical integration
an = []
bn = []
for n in range(1, num_terms + 1):
    an_n = (2 / L) * np.trapz(trapezoid(x_values) * np.cos(2 * n * np.pi * x_values / L ), x_values)
    bn_n = (2 / L) * np.trapz(trapezoid(x_values) * np.sin(2 * n * np.pi * x_values / L), x_values)
    an.append(an_n)
    bn.append(bn_n)

# Create a function to calculate the Fourier series approximation for a given x
def fourier_series(x, num_terms):
    series_sum = a0  # Initialize with a0
    for n in range(1, num_terms + 1):
        series_sum += an[n - 1] * np.cos(2 * n * np.pi * x / L ) + bn[n - 1] * np.sin(2 * n * np.pi * x / L)
    return series_sum

# Generate x values for 5 periods
x_values = np.linspace(- 2 * L , 2 * L, 1500)

# Evaluate the Fourier series for the given x values
approximation = [fourier_series(x, num_terms) for x in x_values]

# Plot the original function and the Fourier series approximation

plt.figure(figsize=(10, 6))
plt.plot(x_values, per_trapezoid(x_values,5), label="Original Function", color='blue')
plt.plot(x_values, approximation, label=f'Fourier Series (n={num_terms})', color='red')
plt.legend()
plt.title('Fourier Series Approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## Second example: step function from the worksheet 8

```python
def step(x):
    return np.piecewise(x, [(x >=0) & (x < 1), (x >= 1) & (x < 2)], [lambda x: 1, lambda x: 0])

L = 2

def per_step(x, num):
    value = step(x) # initialize with the unshifted value
    for n in range(1, num + 1):
        value += step(x - n * L) + step(x + n * L)
    return value

num = 6

x = np.linspace(-2, 4, 600)    
y = per_step(x,num)
plt.plot(x,y)
plt.show()
```

```python
# Number of terms in the Fourier series
num_terms = 100

# Generate x values for one period
x_values = np.linspace(0, L, 600)

# These arrays hold the Fourier series coefficients
a = [0 for _ in range(num_terms + 1)]
b = [0 for _ in range(num_terms + 1)]

a[0] = (1 / L) * np.trapz(step(x_values), x_values)  # get a0 using numerical integration

for n in range(1, num_terms + 1):
    a[n] = (2 / L) * np.trapz(step(x_values) * np.cos(2 * n * np.pi * x_values / L ), x_values)
    b[n] = (2 / L) * np.trapz(step(x_values) * np.sin(2 * n * np.pi * x_values / L), x_values)

# Create a function to calculate the Fourier series approximation for a given x
def fourier_series(x, num_terms):
    series_sum = a[0]  # Initialize with a0
    for n in range(1, num_terms + 1):
        series_sum += a[n] * np.cos(2 * n * np.pi * x / L ) + b[n] * np.sin(2 * n * np.pi * x / L)
    return series_sum

# Generate x values for 5 periods
x_values = np.linspace(- 2 * L , 2 * L, 1600)

# Evaluate the Fourier series for the given x values
approximation = [fourier_series(x, num_terms) for x in x_values]

# Plot the original function and the Fourier series approximation

plt.figure(figsize=(10, 6))
plt.plot(x_values, per_step(x_values,num), label="Original Function", color='blue')
plt.plot(x_values, approximation, label=f'Fourier Series (n={num_terms})', color='red')
plt.legend()
plt.title('Fourier Series Approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## Even periodic extension of a function

```python
def f(x):
    return np.exp(x)
    
x = np.linspace(-4, 4, 900) 
y = f(x)
plt.plot(x,y)
plt.show()
```

Clearly, this function is not a periodic function. Let's assume we're asked to write it in the form  
$ f(x) = a_0 + \sum_{n=1}^{\infty} a_n \cos{ n\pi x\over a}$ for all $x\in [0,a)$.

Let me define the function $f_e(x) =  a_0 + \sum_{n=1}^{\infty} a_n \cos{ n\pi x\over a}$ for all $x$.

Clearly, $f_e(x)$ is a periodic function (sum of periodic terms), and it is even $f_e(-x)=f_e(x)$ because it only has cosines in its Fourier series. Furthermore, the period must be $L=2a$, so that the usual argument ${2n \pi x\over L}={ n \pi x\over a}$. Finally, $f_e(x) = f(x)$ for all $x\in [0,a]$, by definition.

This means that we know $f_e(x)$ for a full period $L=2a$, namely $f_e(x) = \exp(x)$ for $0< x< a$ and $f_e(x)=f_e(-x)= \exp(-x)$ for $-a < x < 0$. This fully determines $f_e(x)$ and so we can calculate its Fourier series as usual. With the periodicity interval chosen as $-L/2, L/2$ and using the fact that the integrand is even, we can reduce the integrals to half the interval:
$$ a_0 = {1\over L}\int_{-{L\over 2}}^{L\over 2}f_e(x)= {1\over a}\int_{0}^a f(x)$$
and for all $n\ge 1$:
$$a_n = {2\over L}\int_{-{L\over 2}}^{L\over 2}f_e(x)\cos{2 n\pi x\over L}= {2\over a}\int_{0}^{a}f(x)\cos{n\pi x\over a}$$

Let's check that this works. We need to specify the value of a, let's say we set it to 3.

```python
a = 2
L = 2 * a

# Number of terms in the Fourier series
num_terms = 10

# Generate x values for one period
x_values = np.linspace(0, a, 300)

# Calculate the Fourier series coefficients
ac = [0 for _ in range(num_terms + 1)]
ac[0] = (1 / a) * np.trapz(f(x_values), x_values)  # get a0 using numerical integration

for n in range(1, num_terms + 1):
    ac[n] = (2 / a) * np.trapz(f(x_values) * np.cos(n * np.pi * x_values / a ), x_values)

# Create a function to calculate the Fourier series approximation for a given x
def fourier_series(x, num_terms):
    series_sum = ac[0]  # Initialize with a0
    for n in range(1, num_terms + 1):
        series_sum += ac[n] * np.cos(2 * n * np.pi * x / L ) 
    return series_sum

# Generate x values for 5 periods
x_values = np.linspace(- 2 * L , 2 * L, 1500)

# Evaluate the Fourier series for the given x values
approximation = [fourier_series(x, num_terms) for x in x_values]

# Plot the original function and the Fourier series approximation

plt.figure(figsize=(10, 6))
plt.plot(x_values, f(x_values), label="Original f(x)=exp(x)", color='blue')
plt.plot(x_values, approximation, label=f'Even periodic extension (a={a},n={num_terms})', color='red')
plt.legend()
plt.title('Fourier Series Approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0,10)
plt.grid(True)
plt.show()
```

## Odd periodic extension of a function

```python
def f(x):
    return np.exp(x)
    
x = np.linspace(-4, 4, 900) 
y = f(x)
plt.plot(x,y)
plt.show()
```

Clearly, this function is not a periodic function. Let's assume we're asked to write it in the form  
$ f(x) = \sum_{n=1}^{\infty} b_n \sin{ n\pi x\over a}$ for all $x\in [0,a)$.

Let me define the function $f_o(x) =  \sum_{n=1}^{\infty} b_n \sin{ n\pi x\over a}$ for all $x$.

Clearly, $f_o(x)$ is a periodic function (sum of periodic terms), and it is odd $f_o(-x)=-f_o(x)$ because it only has sines in its Fourier series. Furthermore, the period must be $L=2a$, so that the usual argument ${2n \pi x\over L}={ n \pi x\over a}$. Finally, $f_o(x) = f(x)$ for all $x\in [0,a]$, by definition.

This means that we know $f_o(x)$ for a full period $L=2a$, namely $f_o(x) = \exp(x)$ for $0< x< a$ and $f_o(x)=-f_o(-x)= -\exp(-x)$ for $-a < x < 0$. This fully determines $f_o(x)$ and so we can calculate its Fourier series as usual. With the periodicity interval chosen as $-L/2, L/2$ and using the fact that the integrand is even, we can reduce the integrals to half the interval, so that for all $n\ge 1$:
$$b_n = {2\over L}\int_{-{L\over 2}}^{L\over 2}f_o(x)\sin{2 n\pi x\over L}= {2\over a}\int_{0}^{a}f(x)\sin{n\pi x\over a}$$

Let's check that this works. We need to specify the value of a, let's say we set it to 2.

```python
a = 2
L = 2 * a

# Number of terms in the Fourier series
num_terms = 500

# Generate x values for one period
x_values = np.linspace(0, a, 3000)

# Calculate the Fourier series coefficients
b = [0 for _ in range(num_terms + 1)]
for n in range(1, num_terms + 1):
    b[n] = (2 / a) * np.trapz(f(x_values) * np.sin(n * np.pi * x_values / a ), x_values)
    bn.append(bn_n)

# Create a function to calculate the Fourier series approximation for a given x
def fourier_series(x, num_terms):
    series_sum = 0  # Initialize 
    for n in range(1, num_terms + 1):
        series_sum += b[n] * np.sin(2 * n * np.pi * x / L ) 
    return series_sum

# Generate x values for 5 periods
x_values = np.linspace(- 2 * L , 2 * L, 1500)

# Evaluate the Fourier series for the given x values
approximation = [fourier_series(x, num_terms) for x in x_values]

# Plot the original function and the Fourier series approximation

plt.figure(figsize=(10, 6))
plt.plot(x_values, f(x_values), label="Original f(x)=exp(x)", color='blue')
plt.plot(x_values, approximation, label=f'Odd periodic extension (a={a},n={num_terms})', color='red')
plt.legend()
plt.title('Fourier Series Approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10,10)
plt.grid(True)
plt.show()
```
